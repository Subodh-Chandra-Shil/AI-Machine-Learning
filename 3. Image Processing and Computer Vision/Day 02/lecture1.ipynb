## OpenCV

```python
import cv2
image = cv2.imread('p1.jpg')
# image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR) # color space conversion 
cv2.imshow('Loaded image', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# splitting into three channels

blue, green, red = cv2.split(image)
rgb_normal = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(6, 10))

plt.subplot(2, 2, 1)
plt.imshow(rgb_normal)

plt.subplot(2, 2, 2)
plt.imshow(red)

plt.subplot(2, 2, 3)
plt.imshow(green)

plt.subplot(2, 2, 4)
plt.imshow(blue)
```
